Дополнительная часть (необязательная)
Улучшить скрипт так, чтобы он анализировал не только preview-информацию статьи, но и весь текст статьи целиком.

Для этого потребуется получать страницы статей и искать по тексту внутри этой страницы.

Итоговый датафрейм формировать со столбцами: <дата> - <заголовок> - <ссылка> - <текст_статьи>

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import time
```

```python
import pandas as pd
```

```python
res = requests.get('https://habr.com/ru/all/')
res
```

```python
KEYWORDS = ['python', 'парсинг', 'Максим']
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
posts = soup.find_all('article', class_='post')
posts
```

```python
all_links = []
for post in posts:
    post_url = post.find('a', class_ = 'post__title_link').get('href')
    all_links.append(post_url)
    if not post_url:
        continue
all_links    
```

```python
posts_with_keywords = pd.DataFrame()
for link in all_links:
    soup = BeautifulSoup(requests.get(link).text, 'html.parser')
#     time.sleep(0.3)
#     text = soup.find('div', class_='post__text ').text
    for keyword in KEYWORDS:
        if keyword in soup.find('div', class_='post__text').text:
            title = soup.find('span', class_='post__title-text').text
            date = pd.to_datetime(soup.find('span', class_='post__time').get('data-time_published'), dayfirst=True).date()
            text = soup.find('div', class_='post__text').text
            row = {'date': date, 'title': title, 'text': text}
            posts_with_keywords = pd.concat([posts_with_keywords, pd.DataFrame([row])]) 
posts_with_keywords
        
```

```python
res_avast = requests.post(url, json = params_post, headers = headers_post)
res_json = res_avast.json()
res_json
```

```python
# находим id взломов для каждой почты (создаем словарm, где ключ - это почта, а значение - id взломов этой почты)

result_email_breaches_id = {}
for key, value in res_json['summary'].items():
    for i in value:
        email_breaches_id = {key: value['breaches'] }
        result_email_breaches_id.update(email_breaches_id)
#         result_email_breaches_id | email_breaches_id
result_email_breaches_id

    
```

```python
result_brench_data = pd.DataFrame()
for key, value in result_email_breaches_id.items():
#     почта
#     дата утечки
#     источник утечки
#     описание утечки
    mail = key
    for i in value:
        date_of_leak =  res_json['breaches'][str(i)]['publishDate']
        source_of_leak = res_json['breaches'][str(i)]['site']
        leak_description = res_json['breaches'][str(i)]['description']
        row = {'почта': mail, 'дата утечки': date_of_leak, 'источник утечки': source_of_leak, 'описание утечки': leak_description}
        result_brench_data = pd.concat([result_brench_data, pd.DataFrame([row])])

result_brench_data = result_brench_data.reset_index().drop(['index'],axis=1)
result_brench_data
```
