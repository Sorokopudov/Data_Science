## Attribute Information:
1:  surgery?
          1 = Yes, it had surgery
          2 = It was treated without surgery

  2:  Age
          1 = Adult horse
          2 = Young (< 6 months)

  3:  Hospital Number
          - numeric id
          - the case number assigned to the horse
            (may not be unique if the horse is treated > 1 time)

  4:  rectal temperature
          - linear
          - in degrees celsius.
          - An elevated temp may occur due to infection.
          - temperature may be reduced when the animal is in late shock
          - normal temp is 37.8
          - this parameter will usually change as the problem progresses
               eg. may start out normal, then become elevated because of
                   the lesion, passing back through the normal range as the
                   horse goes into shock
  5:  pulse
          - linear
          - the heart rate in beats per minute
          - is a reflection of the heart condition: 30 -40 is normal for adults
          - rare to have a lower than normal rate although athletic horses
            may have a rate of 20-25
          - animals with painful lesions or suffering from circulatory shock
            may have an elevated heart rate

  6:  respiratory rate
          - linear
          - normal rate is 8 to 10
          - usefulness is doubtful due to the great fluctuations

  7:  temperature of extremities
          - a subjective indication of peripheral circulation
          - possible values:
               1 = Normal
               2 = Warm
               3 = Cool
               4 = Cold
          - cool to cold extremities indicate possible shock
          - hot extremities should correlate with an elevated rectal temp.

  8:  peripheral pulse
          - subjective
          - possible values are:
               1 = normal
               2 = increased
               3 = reduced
               4 = absent
          - normal or increased p.p. are indicative of adequate circulation
            while reduced or absent indicate poor perfusion

  9:  mucous membranes
          - a subjective measurement of colour
          - possible values are:
               1 = normal pink
               2 = bright pink
               3 = pale pink
               4 = pale cyanotic
               5 = bright red / injected
               6 = dark cyanotic
          - 1 and 2 probably indicate a normal or slightly increased
            circulation
          - 3 may occur in early shock
          - 4 and 6 are indicative of serious circulatory compromise
          - 5 is more indicative of a septicemia

## Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

Значения столбцов 1, 2, 7, 8, 9 являются категориальными - рассчитывать для них выбросы не имеет смысла. Проверим оишь что бы все значение были среди категорий для данного параметра (все значения соответствуют заявоенным категориям.

Найдем выбросы для столбцов [4, Temperature], 	[5, Pulse],	[6, Respiratory rate]

#### Рассмотрим столбец [4, Temperature]
Нормальное значение температуры (здоровое животное) - 37,8
Минимальное значение - 35,4
Максимальное - 40,8

Экстримальные значение мин и макс находятся в разумных пределах. Считаем, что выбрасов в данном столбце нет.

#### Рассмотрим столбец [5, Pulse]
Нормальное значение 30-40 ударов.
У молодых особей нормальный пульс 70 - 120 ударов в минуту.
Минимальное значение 30 - не вызывает подозрений, очевидно не выброс.
Максимальное значение 184 - тоже находится в разумных пределах, т.е. это не нереальный показатель. Веротяно повышенный пульс говорит о наличии заболевания. 
Я считаю, что в данном столбце тоже нет очевыидных выбросов. Но тем не менее найдем выбросы через межквартльный размах и проанализируем их:


```python
q1_5 = my_horse_colic_database_2['_5_Pulse'].quantile(0.25)
q3_5 = my_horse_colic_database_2['_5_Pulse'].quantile(0.75)
iqr_5 = q3_5 - q1_5
lower_bound_5 = q1_5 - (1.5 * iqr_5) 
upper_bound_5 = q3_5 + (1.5 * iqr_5)
print(q1_5, q3_5, iqr_5, lower_bound_5, upper_bound_5)
remove_outliers = my_horse_colic_database_2[my_horse_colic_database_2['_5_Pulse'].between(lower_bound_5, upper_bound_5, inclusive=True)]
# remove_outliers
pd.concat([my_horse_colic_database_2, remove_outliers]).drop_duplicates(keep=False)
```

По результатам мы видим, что повышеные значения пульса встречаются только у молодых особей. У жеребят нормальный пульс 70 - 120 ударов в минуту. Веротяно повышенная частота пульса говорит о болезни этих животных или о возможных физ нагрузказ перед измерением. Пологаю, что нет основыний считать данные значения выбросами.

#### Рассмотрим столбец [6, Respiratory rate] (частота дыхания)

Согласно описанию к данным:
Частота дыхания
- линейный
- нормальная оценка от 8 до 10
- полезность сомнительна из-за больших колебаний

Минимальное значение данного параметра 8 - находится нормальном диапозоне.
Максимальное значение 96 - вероятнее всего является выбросом.

Произведем рассчеты и проанализируем результаты:

```python
q1_6 = my_horse_colic_database_2['_6_Respiratory_rate'].quantile(0.25)
q3_6 = my_horse_colic_database_2['_6_Respiratory_rate'].quantile(0.75)
iqr_6 = q3_6 - q1_6
lower_bound_6 = q1_6 - (1.5 * iqr_6) 
upper_bound_6 = q3_6 + (1.5 * iqr_6)
print(q1_6, q3_6, iqr_6, lower_bound_6, upper_bound_6)
remove_outliers_6 = my_horse_colic_database_2[my_horse_colic_database_2['_6_Respiratory_rate'].between(lower_bound_6, upper_bound_6, inclusive=True)]
# remove_outliers
pd.concat([my_horse_colic_database_2, remove_outliers_6]).drop_duplicates(keep=False)
```

Справочная инфомрация:
Нормальная частота дыхания у взрослых лошадей составляет от 8 до 12 вдохов в минуту, у новорожденных жеребят - от 60 до 80 вдохов в минуту. Жеребята в первые недели жизни сохраняют дыхательные нормы в пределах 20-40 вдохов в минуту.

Большая часть высоких значений данного столбца относится к молодым особям (до 6 месяцев), для которых характерно болеее частое дыхание.   
Также у особей с частым дыханием показатель частоты пульса тоже высокий. Что подтверждает правдивость этих данных.
Считаю данные выбросы укладываются в реальную картину и не требуют корректировки.
