## Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
# five_star - датафрейм, строки в котором только с оценкой 5
five_star = ratings[ratings['rating'] == 5.0]
five_star
```

```python
# best_five_star - вспомогательный датафрейм, в котором хранится индекс фильма с нибольшим колчеством оценок 5 (index) и количество таких оценок(movieid)
best_five_star = five_star['movieId'].value_counts().head(1).reset_index() 
best_five_star
```

```python
movies.loc[movies['movieId'] == best_five_star.iloc[0, 0]]
```

```python
# красивый вывод результата
best_movie = movies.loc[movies['movieId'] == best_five_star.iloc[0, 0]]
print(f"Фильму {best_movie.iloc[0, 1]} было поставлено больше всего оценок '5' в количестве {best_five_star.iloc[0, 1]}")
```

## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
filter_bribaltika = data[
    (   (data['country'] == 'Lithuania') | (data['country'] == 'Latvia')  |  (data['country'] == 'Estonia')    ) & 
    (   (data['year'] < 2011) & (data['year'] > 2004)    ) &
    (   (data['category'] == 4) | (data['category'] == 12)  |  (data['category'] == 21)    )  &
    (    data['quantity'] > 0    )

    ]
```

```python
filter_bribaltika['quantity'].sum() 
```

```python
print(f"Суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года: {filter_bribaltika['quantity'].sum()} ")
```
