1.Переведите содержимое файла purchase_log.txt в словарь purchases вида:
{‘1840e0b9d4’: ‘Продукты’, …}

```python
import json
```

```python
i = 0
d = {}
with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    for line in f:
        if i != 0:
            line = line.strip()
            dict_ = json.loads(line)
            d[dict_["user_id"]] = dict_["category"]        
        i += 1
#         if i > 100:     # на моем ПК весь файл не обрабатывается, поэтому оставил это условие 
#                 break 
print(d)
```

2.Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки (если покупка была, сам файл visit_log.csv изменять не надо). Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.
Учтите условия на данные:
содержимое purchase_log.txt помещается в оперативную память компьютера
содержимое visit_log.csv - нет; используйте только построчную обработку этого файла

```python
with open('visit_log.csv', 'r') as f_read:
    with open('funnel.csv', 'w') as f_write:
        f_write.write('user_id2 , sourse , category\n')
        for i , line in enumerate(f_read):
            if i != 0: 
                user_id2 , sourse = line.strip().split(',')
                if user_id2 in d.keys(): 
                    category = d[user_id2]
                    row_final = f'{user_id2},{sourse},{category}\n'
                    f_write.write(row_final)
                
l = open('funnel.csv', 'r')
```
